```python
import os
import numpy as np
import pandas as pd

os.chdir('/mnt/d/SoftWare/PycharmProjects/AstroCLIP-main')

csv_path = 'dsm/fits/DR16Q_multimodal_dataset.csv'

EXCLUDE_FIELDS = {
    'OBJID', 'THING_ID', 'RA', 'DEC', 'RUN_NUMBER', 'CAMCOL_NUMBER',
    'FIELD_NUMBER', 'PLATE', 'MJD', 'FIBERID', 'ZWARNING', 'SN_MEDIAN_ALL'
}

BANDS = ['U', 'G', 'R', 'I', 'Z']

def analyze_csv_statistics(csv_path):
    """分析CSV文件中数值字段的统计信息"""

    if not os.path.exists(csv_path):
        print(f"错误: 找不到文件 {csv_path}")
        return

    print(f"正在分析: {csv_path}\n")

    df = pd.read_csv(csv_path)
    total_rows = len(df)
    print(f"总行数: {total_rows:,}\n")

    results = {}

    for col_name in df.columns:
        if col_name in EXCLUDE_FIELDS:
            continue

        col_data = df[col_name]

        if not np.issubdtype(col_data.dtype, np.number):
            continue

        valid_data = col_data.dropna()

        if len(valid_data) > 0:
            results[col_name] = compute_statistics(valid_data.values)

    print_results(results)


def compute_statistics(arr):
    """计算统计量"""
    return {
        'mean': np.mean(arr),
        'var': np.var(arr),
        'min': np.min(arr),
        'max': np.max(arr),
        'p1': np.percentile(arr, 1),
        'p5': np.percentile(arr, 5),
        'p95': np.percentile(arr, 95),
        'p99': np.percentile(arr, 99),
        'count': len(arr)
    }


def print_results(results):
    """格式化输出结果"""
    print("=" * 110)
    print(f"{'Field':<20} {'Mean':>10} {'Variance':>12} {'Min':>10} {'Max':>10} {'1%':>8} {'5%':>8} {'95%':>8} {'99%':>8}")
    print("=" * 110)

    field_order = ['Z']
    for base in ['PSFMAG', 'PSFMAGERR', 'EXTINCTION']:
        for band in BANDS:
            field_order.append(f"{base}_{band}")

    for field_name in field_order:
        if field_name not in results:
            continue

        stats = results[field_name]

        print(f"{field_name:<20} {stats['mean']:>10.4f} {stats['var']:>12.4f} "
              f"{stats['min']:>10.4f} {stats['max']:>10.4f} "
              f"{stats['p1']:>8.4f} {stats['p5']:>8.4f} "
              f"{stats['p95']:>8.4f} {stats['p99']:>8.4f}")

    print("=" * 110)
    print(f"\n共分析 {len(results)} 个字段")
```

```python


analyze_csv_statistics(csv_path)
```
